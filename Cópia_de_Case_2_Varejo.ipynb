```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP','Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas[vendas['Preço'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'].fillna('MS',inplace=True)
vendas.info()
```

```python
media_preco = vendas['Preço'].mean()

vendas['Preço'].fillna('media_preco',inplace=True)

vendas.info()
```

```python
vendas.head()
```

```python
vendas[['Preço','Preço_com_frete']].describe()
```

```python
vendas['Preço'] = pd.to_numeric(vendas['Preço'], errors='coerce')
vendas['Preço_com_frete'] = pd.to_numeric(vendas['Preço_com_frete'], errors='coerce')
```

```python
preco_errado = vendas.query('Preço > Preço_com_frete')
preco_correto = vendas.query('Preço < Preço_com_frete')
```

```python
preco_correto.describe()
```

```python
preco_correto.head()
```

```python
preco_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
preco_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
#Média preço com frete por nome de departamento

preco_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index()
```

```python
#Quantidade vendas por DATA

preco_correto.groupby('Data').idcompra.nunique().sort_values(ascending=False)
```

```python
#criar nova coluna com os meses dentro da tabela preco_correto com base nos dados da coluna data
preco_correto ['mes'] = preco_correto['Data'].dt.month
preco_correto.head()
```

```python
#Quantidade vendida POR MÊS

preco_correto.groupby('mes').idcompra.nunique().sort_values(ascending=False)
```

```python
#importando tabela clientes

cliente = pd.read_excel('/content/cliente_varejo.xlsx')
cliente.head()
```

```python
#analisando algumas colunas (idade, renda)
cliente[['idade','renda']].describe()
```

```python
#verificando melhor a tabela (para caso haja algum dado nulo)
cliente.info()
```

```python
#transformando renda -- de int64 para float

cliente = cliente.astype({'renda' : 'float'})
cliente.info()
```

```python
#fazendo join de ambas tabelas (vendas e clientes)

venda_cliente = preco_correto.merge(cliente,how='left',on='cliente_Log')
venda_cliente.head()
```

```python
#AGRUPANDO MÉDIA DE RENDA POR CANAL DE VENDA
agg_idcanal_renda = round(venda_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
#AGRUPANDO MÉDIA DE IDADE POR BANDEIRA
agg_idade_bandeira = round(venda_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
#IMPORTANDO BIBLIOTECA NOVA -- MATPLOTLIB.PYPLOT
import matplotlib.pyplot as plt
```

```python
#GRÁFICO DE BARRAS PLT.BAR DE MÉDIA DE IDADE POR BANDEIRA
fig = plt.figure(figsize=(10,8))
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'],color = 'green')
plt.ylabel('Média de idade')
plt.title('Idade média por bandeira', fontsize = 18, loc= 'center',fontweight='bold',color='black')
```

```python
#GRÁFICO DE BARRAS PLT.BAR DE MÉDIA DE RENDA POR CANAL DE VENDA
fig=plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda['idcanalvenda'],agg_idcanal_renda['renda'],color='green')

plt.ylabel('Renda média')

plt.title('Renda média por canal de venda',loc='center',fontsize=18,fontweight='bold',color='black')
```

```python
#VENDAS POR DATA

venda_por_data = preco_correto.groupby('Data').idcompra.nunique().reset_index()
venda_por_data
```

```python
#GRÁFICO DE LINHA -- VENDAS POR TEMPO(DATA)

fig=plt.figure(figsize=(15,11));

plt.xlabel('Data da venda');
plt.ylabel('Quantidade de vendas');
x = venda_por_data['Data'];
y = venda_por_data['idcompra'];
plt.plot(x,y,color='green',linewidth='2');
```

```python
#GRAFICO LINHA -- MÉDIA DE PREÇO POR NOME DEPARTAMENTO

agg_dept_preco = round(preco_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
#GRAFICO LINHA -- MEDIA PRECO POR NOME DEPARTAMENTO

fig=plt.figure(figsize=(30,20));

plt.bar(agg_dept_preco['Nome_Departamento'],agg_dept_preco['Preço_com_frete'],color='green');

plt.ylabel('Preço médio');

plt.title('Preço médio por departamento',loc='center',fontsize=18,fontweight='bold',color='black');
```
